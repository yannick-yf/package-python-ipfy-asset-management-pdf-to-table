# Output of AWS Textract - Credit Agricole

## Function to unzip output from aws textract
Context: Everything will be located to a dedicated folder. 
For all the zip fil in that folder we will execute the unzip function

```python
import zipfile
import glob

list_file_to_unzip = os.listdir('./pdf_awstextract_output_zip/')
directory_to_extract_to = './awstextract_unzip/'

list_file_to_delete_for_renaming_steps = []

for i in list_file_to_unzip:
    if i == '.DS_Store':
        continue
    else:
        path_to_zip_file = './pdf_awstextract_output_zip/' + i
        print(path_to_zip_file)

        with zipfile.ZipFile(path_to_zip_file, 'r') as zip_ref:
            zip_ref.extractall(directory_to_extract_to)
        
        list_unzipped_files = os.listdir(directory_to_extract_to)
        list_unzipped_files = [x for x in list_unzipped_files if x not in list_file_to_delete_for_renaming_steps]

        for unzipped_file in list_unzipped_files:
        # Absolute path of a file
            old_name = directory_to_extract_to + unzipped_file
            new_name = directory_to_extract_to + i + '_' + unzipped_file

            if os.path.exists(directory_to_extract_to + '/' + new_name):
                continue
            else:
                # Renaming the file
                os.rename(old_name, new_name)
            list_file_to_delete_for_renaming_steps = os.listdir(directory_to_extract_to)

# Read all the file containing table into one csv
aws_texttract_csv_output = pd.concat(
    map(
        pd.read_csv, glob.glob(
            os.path.join(
                './awstextract_unzip/', 
                "*_table-*.csv")
            )
        )
    )
```
